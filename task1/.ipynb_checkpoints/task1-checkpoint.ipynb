```python
# final Code

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.neighbors import KNeighborsClassifier

if __name__ == '__main__':
    # load data
    train_raw = pd.read_csv('introML2018.task1.train.csv')
    test_raw = pd.read_csv('introML2018.task1.test.csv')
    
    # use RandomForestModel
    #model = RandomForestClassifier(n_estimators=25, n_jobs=-1)
    model = KNeighborsClassifier(n_neighbors=5, n_jobs = -1)
    
    for seed in range(10, 11):
        for time in range(0, 1):
            # split
            trainData, testData = train_test_split(train_raw, test_size=0.0, random_state=seed)
            trainData_X = trainData.drop(columns='label')
            trainData_Y = trainData['label']
            testData_X = testData.drop(columns='label')
            testData_Y = testData['label']
            
            # feature selection
            trainData_X = trainData_X.drop(columns='feature_0')
            testData_X = testData_X.drop(columns='feature_0')
            test_raw = test_raw.drop(columns='feature_0')

            # learn
            model.fit(trainData_X, trainData_Y)

            # predict
            test_raw_Y = model.predict(test_raw)
    #        test_Y = model.predict(testData_X)

            # evaluate score
     #       print("Seed =", seed,"-- accuracy_score:", accuracy_score(testData_Y, test_Y))

            # save result
            result = pd.DataFrame()
            result['ID'] = range(1, len(test_raw_Y) + 1)
            result['Label'] = test_raw_Y
            resultName = 'introML2018.task1.predict_' + repr(time+seed*10) + '.csv'
            result.to_csv(resultName, index=False)
```
